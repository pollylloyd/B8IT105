```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#read in the file and name it as a data frame 'df'
df = pd.read_csv('avocado.csv')
'Data set is read'
```

```python
# Convert date to a date type and type to categorical
df['Date'] = pd.to_datetime(df['Date']) 
df['type'] = df['type'].astype('category')
#df['year'] = df['year'].astype('category')
df.dtypes
```

```python
# And Sort by date in ascending date order -
df.sort_values(by='Date', ascending=True)
```

```python
# Basic statistical information
df.describe()
# Note the overall mean is $1.41, the min average price was $0.44 and the max average price was $3.25
```

```python
# Reorder columns - to make more logical order of data
df = df.reindex(columns = ['Date', 'year', 'region', 'type','AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Small Bags', 'Large Bags', 'XLarge Bags', 'Total Bags'])
df
```

```python
df = pd.read_csv('avochange.csv')
df
```

```python
# Cleaning up - Drop the 'Unnamed: 0' column, which isn't useful
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
df_type = df.groupby('type')['Total Volume'].agg(sum)
df_type
```

```python
plt.pie([df_type], explode=(0.1, 0), labels=['Conventional', 'Organic'], colors=['orange', 'green'], 
autopct='%1.1f%%', startangle=150)
title='Total Proportion of all Avocados by Type'
plt.axis('equal')
plt.show()
```

```python
# Alternative visualisation - in absolute figures
df_type.plot(kind = 'bar',color = 'blue',rot=0,figsize=(20,5), title='Total Volume of all Avocados by Type, in tens of Billions')
plt.ylabel('Tens of Billions of Avocados')
plt.xlabel('Type')
plt.show()
#significantly more organic avocados
```

```python
# Show volume by year
df_year = df.groupby('year')['Total Volume'].agg(sum)
df_year

# Note that the figures show an increase 2015 through 2017.  2018 represents only partial year
```

```python
df_year.plot(kind = 'bar',color = 'red',rot=0,figsize=(20,5), title='Total Volume of all Avocados by Year, in Billions')
plt.ylabel('Quantity in Billions')
plt.xlabel('Year')
plt.show()
```

```python
# Some high level Volumes
print('Hass Volumes by Year and Type')
df_type_year = df.groupby(['year','type'])['Total Volume'].agg(sum)
df_type_year

```

```python
# Total volume - show total volume
df['Total Volume'].agg(sum)
```

```python
#Broken down further by Year and Type
df_type_year.plot(kind = 'bar',color = ['orange','green'],rot=45, figsize=(20,5), title='Volume by Year and Type')
plt.ylabel('Quantity in Billions', fontsize = 12)
plt.xlabel('Year and Type', fontsize = 12)
plt.show()
```

```python
# Some high level Price states
print('Hass Prices in USD by Year and Type')
df_price = df.groupby(['year','type'])['AveragePrice'].agg([np.mean])
df_price
#
```

```python
# Turn same data into a pivot table, but with index (rows) as Year and columns split between min, max, mean for the
# different types

df_pivot_price = df.pivot_table(index=['year', 'type'], aggfunc=[np.mean], values = 'AveragePrice')
df_pivot_price
```

```python

plt.figure(figsize = (12, 8))
sns.barplot(x = 'year', y = 'AveragePrice', hue = 'type', data = df, palette = ['orange', 'green'])
plt.title('Average Price of Avocados by Year and Category', fontsize = 14)
plt.ylabel('USD $', fontsize = 14)
plt.xlabel('Year', fontsize = 14)
plt.show()
```

```python
df_ave_trend = df.groupby('Date').mean()
df['Date'] = pd.to_datetime(df['Date']) 
df['type'] = df['type'].astype('category')
df.sort_values(by='Date', ascending=True)
```

```python
df_ave_trend['AveragePrice'].plot(figsize=(20,5), color = 'purple', title = 'Average Price of all Avocados accross all states')
plt.xlabel('')
plt.ylabel('Average Price USD $')
plt.show()
```

```python
# Smooth out the trend line for 20 Data points rolling average
df_ave_trend['AveragePrice'].rolling(20).mean().plot(figsize=(20,5), color = 'purple', title = 'Rolling average (20 data points) Price of all Avocados accross all states')
plt.xlabel('')
plt.ylabel('Average Price USD $')
plt.show()
```

```python
# Group the entire data set by Region and return the mean of the Average Price by year
# Note this is aggregated for both types - conventional and organic
filter_by_region = df.groupby(['region'])['AveragePrice'].mean() #filter_by_region = df.groupby(['region','year'])['AveragePrice'].mean()
```

```python
filter_by_region.plot(kind = 'bar',color = 'purple',rot=90,figsize=(20,5), title='Average Price in USD$ for all Avocados by Region')
plt.ylabel('Ave Price USD$')
plt.show()
```

```python
# Show the Ten Regions with the Highest Average Price
filter_by_region.nlargest(10).plot(kind = 'bar',color = 'purple',rot=45,figsize=(20,5), title='Ten regions with the highest average price in USD$ for all avocados')
plt.ylabel('Ave Price USD$')
plt.show()
```

```python
# Show the Ten Regions with the Lowest Average Price
filter_by_region.nsmallest(10).plot(kind = 'bar',color = 'blue',rot=45,figsize=(20,5), title='Ten regions with the lowest average price in USD$ for all avocados')
plt.ylabel('Ave Price USD$')
plt.show()
# Largely regions near to the growing regions of California, Florida and Mexico
```
